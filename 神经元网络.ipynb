```python
import pandas as pd

train = pd.read_csv('train.csv',encoding='gbk')
test = pd.read_csv('test.csv',encoding='utf-8')

print(train.head(5))
print('--------------------------')
print(test.head(5))
```

```python
print(train.isnull().sum())
print('-------------------')
print(test.isnull().sum())
```

```python
orient_mapping={
    '东':0.1,'东南':0.2,'南':0.3,'西南':0.4,
    '西':0.5,'西北':0.6,'北':0.7,'东北':0.8,
}
```

```python
train['房屋朝向']=train['房屋朝向'].map(orient_mapping)
test['房屋朝向']=test['房屋朝向'].map(orient_mapping)
```

```python
train['小区房屋出租数量'] = train['小区房屋出租数量'].fillna(train['小区房屋出租数量'].mean())
test['小区房屋出租数量'] = test['小区房屋出租数量'].fillna(test['小区房屋出租数量'].mean())
```

```python
train['区'] = train['区'].fillna(train['区'].mean())
test['区'] = test['区'].fillna(test['区'].mean())
```

```python
train['位置'] = train['位置'].fillna(train['位置'].mean())
test['位置'] = test['位置'].fillna(test['位置'].mean())
```

```python
#按照最小值填充=1
train['地铁线路'] = train['地铁线路'].fillna(1)
test['地铁线路'] = test['地铁线路'].fillna(1)
```

```python
#按照最小值填充=1
train['地铁站点'] = train['地铁站点'].fillna(0)
test['地铁站点'] = test['地铁站点'].fillna(0)
```

```python
#按照最大值填充=1
train['距离'] = train['距离'].fillna(1)
test['距离'] = test['距离'].fillna(1)
```

```python
train.drop ('时间', axis=1,inplace=True)
test.drop ('时间', axis=1,inplace=True)
```

```python
train.drop ('居住状态', axis=1,inplace=True)
test.drop ('居住状态', axis=1,inplace=True)
```

```python
train.drop ('出租方式', axis=1,inplace=True)
test.drop ('出租方式', axis=1,inplace=True)
```

```python
train.drop ('装修情况', axis=1,inplace=True)
test.drop ('装修情况', axis=1,inplace=True)
```

```python
train.drop ('房屋朝向', axis=1,inplace=True)
test.drop ('房屋朝向', axis=1,inplace=True)
```

```python
test.drop('id',axis=1,inplace=True)
```
